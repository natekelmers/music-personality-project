```python
def clean_row(row):
    
    # remove formatting created by reading a list as a string
    row = row.strip("[").strip("]")
    
    # restore list property by splitting on (', '), (", '), (', "), and (", ") with regex 
    row = re.split("[\"\'], [\"\']", row)
    
    # removes any stray string delimiters from each title, adds clean titles to a new list
    new_row = []
    
    for title in row:
        title = title.strip("'")
        title = title.strip('"')
        
        new_row.append(title)
        
    return new_row
```

```python
mbti['titles_all'] = mbti.titles_all.apply(lambda x: clean_row(x))
```

```python
lyrics_all = []

for row in titles_all[35001: ]:
    
    row_lyrics = []
    
    for title in row:
        
        lyrics = ""
        
        if title:
            
            driver.get(r'https://www.google.com/')
            
            input_searchbox = driver.find_element_by_xpath(r'//*[@id="lst-ib"]')
            input_searchbox.send_keys(title + " " + "lyrics", Keys.ENTER)

            html = driver.page_source
            soup = BeautifulSoup(html)

            lyricsoup = soup.find("div", class_ = "SALvLe farUxc mJ2Mod")

            if lyricsoup:
                if lyricsoup.find("div", class_ = "rGtH5c"):
                    lyricsoup.find("div", class_ = "rGtH5c").decompose()
                    lyricsoup.find("div", class_ = "xpdxpnd kno-fb-ctx R5GB6b v2b5e").decompose()
                    lyrics = lyricsoup.getText(" ")
        
        row_lyrics.append(lyrics)
    
    lyrics_all.append(row_lyrics)
```

```python
lyrics_df = pd.DataFrame(lyrics_all)
lyrics_df.to_csv('C:\Users\Nathan.y.kelmers\Desktop\GA\FINAL PROJECT\data\mbti_lyrics_35001-39996.csv',
                encoding = 'utf-8')
```
