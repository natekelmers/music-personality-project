```python
import pandas as pd
pd.options.display.max_colwidth = 1000

path = "C:/Users/Nathan.y.kelmers/Desktop/GA/FINAL PROJECT/data/mbti_lyrics_complete.csv"
mbti = pd.read_csv(path, header = 0)

import re

import numpy as np
```

```python
def add_space_for_merge(x):
    if x:
        return x + " "
    else:
        return x
```

```python
mbti['lyrics_1'].fillna(value = '', inplace = True)
mbti['lyrics_2'].fillna(value = '', inplace = True)
mbti['lyrics_3'].fillna(value = '', inplace = True)
mbti['lyrics_4'].fillna(value = '', inplace = True)
mbti['lyrics_5'].fillna(value = '', inplace = True)
mbti['lyrics_6'].fillna(value = '', inplace = True)
mbti['lyrics_7'].fillna(value = '', inplace = True)
mbti['lyrics_8'].fillna(value = '', inplace = True)
mbti['lyrics_9'].fillna(value = '', inplace = True)
mbti['lyrics_10'].fillna(value = '', inplace = True)

mbti['lyrics_1'] = mbti['lyrics_1'].apply(lambda x: add_space_for_merge(x))
mbti['lyrics_2'] = mbti['lyrics_2'].apply(lambda x: add_space_for_merge(x))
mbti['lyrics_3'] = mbti['lyrics_3'].apply(lambda x: add_space_for_merge(x))
mbti['lyrics_4'] = mbti['lyrics_4'].apply(lambda x: add_space_for_merge(x))
mbti['lyrics_5'] = mbti['lyrics_5'].apply(lambda x: add_space_for_merge(x))
mbti['lyrics_6'] = mbti['lyrics_6'].apply(lambda x: add_space_for_merge(x))
mbti['lyrics_7'] = mbti['lyrics_7'].apply(lambda x: add_space_for_merge(x))
mbti['lyrics_8'] = mbti['lyrics_8'].apply(lambda x: add_space_for_merge(x))
mbti['lyrics_9'] = mbti['lyrics_9'].apply(lambda x: add_space_for_merge(x))
mbti['lyrics_10'] = mbti['lyrics_10'].apply(lambda x: add_space_for_merge(x))
```

```python
mbti['lyrics_all'] = (mbti['lyrics_1'] + 
                      mbti['lyrics_2'] + 
                      mbti['lyrics_3'] + 
                      mbti['lyrics_4'] + 
                      mbti['lyrics_5'] + 
                      mbti['lyrics_6'] + 
                      mbti['lyrics_7'] + 
                      mbti['lyrics_8'] + 
                      mbti['lyrics_9'] + 
                      mbti['lyrics_10'])
```

```python
def i_vs_e(x):
    if x != 'Unknown':
        if x[0] == 'I':
            return 0
        if x[0] == 'E':
            return 1
    else:
        return None
```

```python
mbti['i_vs_e'] = mbti['mbti'].apply(lambda x: i_vs_e(x))
```

```python
mbti.drop(['username', 'lyrics_1', 'lyrics_2', 'lyrics_3', 'lyrics_4', 'lyrics_5', 
    'lyrics_6', 'lyrics_7', 'lyrics_8', 'lyrics_9', 'lyrics_10'], axis = 1, inplace = True)  
```

```python
def clean_blanks(x):
    if x == []:
        return None
    elif x == "":
        return None
    else:
        return x
```

```python
mbti['lyrics_all'] = mbti['lyrics_all'].apply(lambda x: clean_blanks(x))
```

```python
mbti.dropna(inplace = True)
mbti
```

```python
# FINALLY CLEAN! LET'S MODEL NOW!
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.utils import resample
from sklearn.metrics import roc_auc_score
```

```python
# Looks like we need to balance these...

mbti['i_vs_e'].value_counts()
```

```python
mbti_majority = mbti[mbti['i_vs_e'] == 0]
mbti_minority = mbti[mbti['i_vs_e'] == 1]
```

```python
maj_downsampled = resample(mbti_majority, replace = False, n_samples = 2760, random_state = 42)

mbti_downsampled = pd.concat([maj_downsampled, mbti_minority])
```

```python
X = mbti_downsampled['lyrics_all']
y = mbti_downsampled['i_vs_e']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
vect = CountVectorizer(ngram_range = (1, 3), stop_words = 'english')

X_train_dtm = vect.fit_transform(X_train)
X_test_dtm = vect.transform(X_test)

nb = MultinomialNB()
nb.fit(X_train_dtm, y_train)

y_pred_class = nb.predict(X_test_dtm)

print(metrics.accuracy_score(y_test, y_pred_class))
```

```python
y_pred_prob = nb.predict_proba(X_test_dtm)[:, 1]
auc_score = metrics.roc_auc_score(y_test, y_pred_prob)
auc_score
```

```python
metrics.confusion_matrix(y_test, y_pred_class)
```

```python
# Making the dataframe that lets us look at false positives and negatives

mbti_eval = pd.DataFrame(X_test)
mbti_eval['_prediction_'] = y_pred_class
mbti_eval['_actual_'] = y_test

mbti_indexes = [i for i in mbti_eval.index]
mbti_eval_titles = [mbti['titles_all'].loc[i] for i in mbti_indexes]
mbti_eval['_titles_'] = mbti_eval_titles
```

```python
# predicted introvert but actually an extrovert listened to that
mbti_eval[(mbti_eval['_prediction_'] == 0) & (mbti_eval['_actual_'] == 1)]

# Obvious wrong predictions: Daft Punk "One More Time", Rolling Stones "Street Fighting Man"
```

```python
# Calculating "introvertedness" and "extrovertedness" of each token (all tokens in vocabulary)
mbti_downsampled_i = mbti_downsampled[mbti_downsampled['i_vs_e'] == 0]
mbti_downsampled_e = mbti_downsampled[mbti_downsampled['i_vs_e'] == 1]

vect.fit(mbti_downsampled['lyrics_all'])
mbti_vocabulary = vect.get_feature_names()

mbti_downsampled_i_dtm = vect.transform(mbti_downsampled_i['lyrics_all'])
mbti_downsampled_e_dtm = vect.transform(mbti_downsampled_e['lyrics_all'])

mbti_i_counts = np.sum(mbti_downsampled_i_dtm.toarray(), axis = 0)
mbti_e_counts = np.sum(mbti_downsampled_e_dtm.toarray(), axis = 0)

mbti_token_counts = pd.DataFrame({'token' : mbti_vocabulary, 'i' : mbti_i_counts, 'e' : mbti_e_counts})

mbti_token_counts['i'] = mbti_token_counts['i'] + 1
mbti_token_counts['e'] = mbti_token_counts['e'] + 1

mbti_token_counts['e_ratio'] = mbti_token_counts['e'] / mbti_token_counts['i']
mbti_token_counts.sort_values('e_ratio', ascending = False)
```

```python
# it looks there are some songs with very repetitive lyrics that are skewing certain tokens (see all of the
# "pumped up kicks" in the "introverted" list and "rollin rollin rollin" in the extroverted list. So, let's see
# if things look better if we only use tokens that appear in many songs
```
