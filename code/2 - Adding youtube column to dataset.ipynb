```python
import pandas as pd
pd.options.display.max_colwidth = 1000

import re

path = "C:/Users/Nathan.y.kelmers/Desktop/GA/FINAL PROJECT/data/mbti_combined.csv"

mbti_columns = ["post", "username", "mbti", "postsoup"]
mbti = pd.read_csv(path, header = 0, names = mbti_columns)
```

```python
posts = [post for post in mbti["postsoup"]]
youtube_urls = []

for post in posts:
    post_parts = re.split("<iframe class\=\"restrain\"|<a href", post)
    links = []
    for part in post_parts:
        if "youtube.com/embed" in part:
            link = re.search("youtube.com\/embed\/(.*)", part)
            if link:
                links.append(link.group(1)[0: 11])
        elif "youtube.com/watch" in part:
            link = re.search("youtube.com\/watch\?v\=(.*)", part)
            if link:
                links.append(link.group(1)[0: 11])
#        else:
#            links.append("None")
    links = ",".join(links)
    youtube_urls.append(links)

mbti["youtube"] = youtube_urls
```

```python
youtube_titles = []

for i in youtube_urls:
    if i:
        results = youtube.videos().list(id = i, part = "snippet").execute()
        titles = [result["snippet"]["title"].encode("utf-8") for result in results.get("items", [])]
        youtube_titles.append(titles)
    else:
        youtube_titles.append("None")
        
mbti["youtube_titles"] = youtube_titles
```
